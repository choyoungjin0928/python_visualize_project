```python
import csv
f = open('seoul1.csv')
data = csv.reader(f)
header = next(data)
print(header)
max_temp = -999
max_date = ''
for row in data:
    if row[-1] == '' :
        row[-1] = -999
    row[-1] = float(row[-1])
    #print(row)
    if max_temp < row[-1]:
        max_temp = row[-1]
        max_date = row[0]
f.close()
print('서울 가장 더웠던 날은 %s이고 %d도 입니다.' %(max_date, max_temp))
```
