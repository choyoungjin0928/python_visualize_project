```python
import csv
f = open('age.csv')  #csv 파일의 속성을 엑셀에서 숫자로 변환 후 저장
data = csv.reader(f)
next(data)
name = input("인구구조가 알고 싶은 지역 이름(읍면동 단위):")
mn = 1
result_name = ''
result = 0
for row in data:
    if name in row[0]:        
        home = np.array(row[3:], dtype=int) / int(row[2])
        #print(home)
f.seek(0)
data = csv.reader(f)
next(data)
for row in data :
#    print(row)    
#    print(int(row[3]), int(row[2]))
    away = np.array(row[3:], dtype=int) / int(row[2]) #비율 계산
#    print(away)
    s = np.sum((home-away)**2)  #인구 분포 비교
#    print(s)
    if s < mn and name not in row[0] :  # 기준값과 결과값 변경
        mn = s
        result_name = row[0]
        result = away
        #print(result)
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(name + '지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(result, label = result_name)
plt.legend()
plt.show()
```
