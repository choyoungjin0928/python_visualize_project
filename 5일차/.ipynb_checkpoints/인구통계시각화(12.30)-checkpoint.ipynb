```python
import csv, matplotlib.pyplot as plt
f = open('age.csv')
data = csv.reader(f)
#print(next(data))
name = input("동네명을 입력하세요: ")
result = []
for row in data:
    if name in row[0]:
        for i in row[3:]:
            print(int(i))
            result.append(int(i))
plt.style.use('ggplot')
plt.plot(result)
print(len(result))
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
file = open('gender.csv')
data = csv.reader(file)
m = []
f = []
for row in data:
    if '신갈' in row[0]:
        for i in row[3:104]:
            m.append(int(i))
        for j in row[106:]:
            f.append(int(j))

plt.title('Family')
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt, csv
file = open('gender.csv')
data = csv.reader(file)
m = []
f = []
for row in data:
    if '신갈' in row[0]:
        for i in row[3:104]:
            m.append(int(i))
        for j in row[106:]:
            f.append(int(j))

plt.title('Family')
plt.scatter(m, f, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
```

```python
import matplotlib.pyplot as plt
import random
x = []
y = []
size = []
for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 200))
    
plt.style.use('ggplot')
plt.scatter(x, y, s = size, c = size, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
