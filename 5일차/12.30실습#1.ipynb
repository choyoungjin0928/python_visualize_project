```python
import matplotlib.pyplot as plt, csv
f = open('seoul1.csv')
data = csv.reader(f)
next(data)
result = []
spr = []
sum = []
fall = []
win = []
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08':
            sum.append(float(row[-1]))
        elif month == '01':
            win.append(float(row[-1]))
        elif month == '04':
            spr.append(float(row[-1]))
        elif month == '09':
            fall.append(float(row[-1]))
        
#plt.boxplot(result)
plt.boxplot([spr, sum, fall, win])
plt.show()
```

```python
import matplotlib.pyplot as plt, csv
f = open('seoul1.csv')
data = csv.reader(f)
next(data)
result = [[],[],[],[],[],[],[],[],[],[],[],[]]
for row in data:
    month = int(row[0].split('-')[1])
    if row[-1] != '' :
        result[int(month)-1].append(float(row[-1]))
#plt.boxplot(result)
plt.boxplot(result)
plt.show()
```
