```python
import matplotlib.pyplot as plt
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)
mx = 0
rate = 0
label = ['유임승차', '유임하차', '무임승차', '무임하차']
for row in data :
    for i in range(4,8) :
        row[i] = int(row[i])
    if row[6] != 0 and (row[4] + row[6]) > 100000:
        rate = row[4] / (row[4] + row[6])    
    if rate > mx :
        mx = rate
        #print(mx, row)
plt.title(row[3] + ' ' + row[1])
plt.pie(row[4:8], labels = label)
plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
plt.show()
#print(row)
#print(mx)
```
